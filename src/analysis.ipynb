```python

plt.figure(figsize=(6,4))

NArr = np.array([10,20,40,80])
energyArr = np.zeros(4)

for i in range(len(NArr)):
    N = NArr[i]
    data = np.loadtxt(f'/home/biscuit/Documents/utd/phys5v48/phys5v48hw/hw2/quantum-eigen/output/outputMultNHarmonic100/eigs_N{N}.txt')
    energyArr[i] = np.average(data,axis=0)

plt.plot(NArr, energyArr, '-')
#plt.xscale('symlog')
plt.yscale('symlog')
plt.xlabel('N')
plt.ylabel('Average Energy (arbitrary units)')
plt.title('Average Energy vs. N')
plt.show()
```

```python
from math import nan


plt.figure(figsize=(6,4))

NArr = np.array([10,20,40,80])

for j in range(0,6):

    energyArr = np.zeros(4)
    
    for i in range(len(NArr)):
        
        if ((i == 0) & (j == 5)):
            energyArr[i] = np.nan
            i += 1
        else:
            N = NArr[i]
            data = np.loadtxt(f'/home/biscuit/Documents/utd/phys5v48/phys5v48hw/hw2/quantum-eigen/output/outputMultNHarmonic100/eigs_N{N}.txt')
            energyArr[i] = data[j]
    
    plt.plot(NArr, energyArr, '--', alpha=1, label="E"+str(j))

plt.legend()
#plt.xscale('symlog')
plt.yscale('symlog')
plt.xlabel('N')
plt.ylabel('Average Energy (arbitrary units)')
plt.title('Average Energy vs. N (Harmonic)')
plt.show()
```
